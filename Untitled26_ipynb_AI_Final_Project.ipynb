```python
# Cell 1: Setup and Data Upload
import pandas as pd
from google.colab import files
import io

# Upload the dataset (you will be prompted to select the CSV file)
print("Please upload the file 'ER_Overload_Seasonal_Weather (1).csv'")
uploaded = files.upload()

# The uploaded file name is 'ER_Overload_Seasonal_Weather (1).csv'
FILE_NAME = "ER_Overload_Seasonal_Weather (1).csv"

# Load the dataset into a DataFrame
if FILE_NAME in uploaded:
    df = pd.read_csv(io.BytesIO(uploaded[FILE_NAME]))
else:
    print(f"Error: Could not find the file {FILE_NAME} in uploaded files.")

print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# Cell 2: Import Visualization Libraries
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cell 3: Data Cleaning and Preparation

# Convert 'Date' column to datetime format and sort
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)

# Check data info
print("Dataset Info:")
df.info()

# Step 4: Add a numerical 'Month' column for seasonal filtering/grouping
df['Month'] = df['Date'].dt.month
```

```python
# Cell 4: Time-Series Plot (ER Visits over time)

print("--- Daily ER Visits Over Time (Interactive Plot) ---")
# Use plotly for an interactive time-series chart
fig = px.line(df, x='Date', y='ER_Visits',
              title='Daily ER Visits Over Time',
              color_discrete_sequence=['#ff4d4d'])
fig.update_layout(xaxis_title="Date", yaxis_title="ER Visits Count")
fig.show()
```

```python
# Cell 5: Histogram of ER Visits

plt.figure(figsize=(10, 6))
# Use Matplotlib/Seaborn for a static histogram
sns.histplot(df['ER_Visits'], bins=20, kde=True, color='#007acc')
plt.title('Histogram of Daily ER Visits', fontsize=16)
plt.xlabel('ER Visits Count', fontsize=12)
plt.ylabel('Frequency (Days)', fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# Cell 6: Seasonal Box Plot (ER Visits by Season)

print("--- ER Visits Distribution by Season (Interactive Box Plot) ---")
# Use plotly for an interactive box plot
fig = px.box(df, x='Season', y='ER_Visits',
             title='ER Visits Distribution by Season',
             color='Season',
             color_discrete_map={'Winter':'#1f77b4', 'Summer':'#ff7f0e',
                                 'Monsoon':'#2ca02c', 'Post-Monsoon':'#d62728'})
fig.update_layout(xaxis_title="Season", yaxis_title="ER Visits Count")
fig.show()
```

```python
# Cell 7: Correlation Heatmap

# Define the variables for correlation analysis
correlation_vars = ['ER_Visits', 'Temperature(°C)', 'Humidity(%)', 'Rainfall(mm)', 'Flu_Cases', 'AirQualityIndex']

# Calculate the correlation matrix
correlation_matrix = df[correlation_vars].corr()

plt.figure(figsize=(10, 8))
# Create the heatmap
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5,
            cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Between ER Visits and Weather/Illness Variables', fontsize=14)
plt.show()
```

```python
# Cell 8: Scatter Plot (ER Visits vs. Flu Cases)

print("--- ER Visits vs. Flu Cases (Interactive Scatter Plot) ---")
fig = px.scatter(df, x='Flu_Cases', y='ER_Visits',
                 title='Relationship Between Flu Cases and ER Visits',
                 trendline="ols", # Add a linear trendline
                 color='Season',
                 hover_data=['Date', 'Temperature(°C)'])
fig.update_layout(xaxis_title="Flu Cases Count", yaxis_title="ER Visits Count")
fig.show()
```

```python
# Cell 9: Prophet Model Data Preparation (ds, y)

# Import Prophet library
from prophet import Prophet

# Create a copy of the DataFrame for forecasting
prophet_df = df.copy()

# The Prophet model requires two columns:
# 1. 'ds' (datetime stamp)
# 2. 'y' (the variable to forecast - ER_Visits)
prophet_df.rename(columns={'Date': 'ds', 'ER_Visits': 'y'}, inplace=True)

# Prophet also allows including extra predictors (Regressors)
# Let's add Flu_Cases and Temperature as extra regressors
REGRESSORS = ['Flu_Cases', 'Temperature(°C)']

print("Prophet-ready data (ds, y, and regressors):")
print(prophet_df[['ds', 'y'] + REGRESSORS].tail())
```

```python
# Cell 10: Monthly Average ER Visits (Advanced Analysis)

# Calculate the average ER visits for each month
monthly_avg_df = df.groupby(df['Date'].dt.to_period('M'))['ER_Visits'].mean().reset_index()
monthly_avg_df['Date'] = monthly_avg_df['Date'].dt.to_timestamp()
monthly_avg_df['Month_Name'] = monthly_avg_df['Date'].dt.strftime('%b')

print("--- Average ER Visits per Month (Bar Chart) ---")
# Use plotly for an interactive bar chart to visualize seasonality
fig = px.bar(monthly_avg_df,
             x='Month_Name',
             y='ER_Visits',
             title='Average ER Visits by Month (Clear Seasonality)',
             color='ER_Visits',
             color_continuous_scale=px.colors.sequential.Plasma,
             labels={'Month_Name': 'Month', 'ER_Visits': 'Average Daily ER Visits'})
fig.update_layout(xaxis={'categoryorder':'array',
                         'categoryarray':['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']})
fig.show()
```

```python
# Cell 11: Training the Prophet Model and Forecasting

# 1. Split data into training and testing (hold out the last 30 days for testing)
TRAIN_SIZE = len(prophet_df) - 30
train_df = prophet_df.iloc[:TRAIN_SIZE]
test_df = prophet_df.iloc[TRAIN_SIZE:]

# 2. Initialize the Prophet model
# Setting seasonality_mode to 'multiplicative' often works well for time series where
# the seasonal amplitude grows with the trend (like ER visits).
model = Prophet(
    seasonality_mode='multiplicative',
    interval_width=0.95 # Confidence Interval (to visualize uncertainty)
)

# 3. Add Regressors
for regressor in REGRESSORS:
    model.add_regressor(regressor)

# 4. Train the model
print("Training Prophet model...")
model.fit(train_df)
print("Prophet model training complete.")

# 5. Create a future dataframe for prediction (e.g., next 60 days)
# We need to extend the future dataframe beyond the training period.
future = model.make_future_dataframe(periods=60, include_history=True)

# For prediction, we need to populate the regressor columns (Flu_Cases and Temperature)
# in the 'future' dataframe, which is a major challenge in real-life forecasting.
# For this simulation, we'll use a simple approach: fill future values with the average
# of the training set's regressor values.
for regressor in REGRESSORS:
    avg_regressor_value = train_df[regressor].mean()
    # Fill the entire future dataframe with the historical average for simplicity
    future[regressor] = future['ds'].apply(lambda x: train_df.loc[train_df['ds'].dt.date == x.date(), regressor].iloc[0] if x.date() in train_df['ds'].dt.date.values else avg_regressor_value)

# 6. Make the prediction
forecast = model.predict(future)

print("\nLast 5 predicted values:")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
# Cell 12: Visualize Forecast Results

print("--- Prophet Model Forecast Visualization ---")

# Plot the forecast: history, prediction, and uncertainty
fig = model.plot(forecast, xlabel='Date', ylabel='Predicted ER Visits')
plt.title('Prophet Forecast of Daily ER Visits', fontsize=16)
plt.show()

# Plot the components (Trend, Weekly Seasonality, Yearly Seasonality, and Regressors)
# This visualization helps explain WHY the model predicts certain peaks.
print("--- Prophet Model Components (Trend and Seasonality) ---")
fig_components = model.plot_components(forecast)
plt.show()

# Optional: Evaluate Model Accuracy (Mean Absolute Error on Test Set)
from sklearn.metrics import mean_absolute_error

# Extract actual vs. predicted for the test period
# Note: We need to merge the actual test data with the forecast results
comparison_df = forecast.set_index('ds')[['yhat']].join(prophet_df.set_index('ds')[['y']]).dropna()
test_comparison = comparison_df.iloc[-30:] # Focus on the last 30 days (test set)

# Calculate MAE
mae = mean_absolute_error(test_comparison['y'], test_comparison['yhat'])

print(f"\nMean Absolute Error (MAE) on the last 30 days of historical data: {mae:.2f}")
print("This MAE value indicates the average magnitude of error in your daily predictions.")
```
